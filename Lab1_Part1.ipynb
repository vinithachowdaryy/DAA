```python
import time
import matplotlib.pyplot as plt

# Iterative approach to find sum of first N natural numbers
def iterative_sum(N):
    total = 0
    for i in range(1, N + 1):
        total += i
    return total

# Recursive approach to find sum of first N natural numbers
def recursive_sum(N):
    if N == 0:
        return 0
    else:
        return N + recursive_sum(N - 1)

# Function to measure execution time
def measure_time(func, N):
    start_time = time.time()
    func(N)
    end_time = time.time()
    return end_time - start_time

# Range of N values to test
N_values = range(10, 501, 10)  # Testing from N=10 to N=500, stepping by 10

# Lists to hold the execution times
iterative_times = []
recursive_times = []

# Measure execution times for each N value
for N in N_values:
    iterative_times.append(measure_time(iterative_sum, N))
    recursive_times.append(measure_time(recursive_sum, N))

# Convert the measured times to milliseconds for better granularity
iterative_times_ms = [time_ms * 1000 for time_ms in iterative_times]
recursive_times_ms = [time_ms * 1000 for time_ms in recursive_times]

# Plotting the results in milliseconds
plt.figure(figsize=(10, 6))
plt.plot(N_values, iterative_times_ms, label='Iterative', marker='o')
plt.plot(N_values, recursive_times_ms, label='Recursive', marker='x')
plt.xlabel('N')
plt.ylabel('Time taken (milliseconds)')
plt.title('Comparing Performance of Iterative vs Recursive Summation (in ms)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import time
import random
import matplotlib.pyplot as plt

# Function to generate a random array of integers
def generate_random_array(size=10000, min_val=1, max_val=1000):
    return [random.randint(min_val, max_val) for _ in range(size)]

# Linear search function
def linear_search(arr, search_key):
    for i in range(len(arr)):
        if arr[i] == search_key:
            return i
    return -1

# Binary search function
def binary_search(arr, search_key):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == search_key:
            return mid
        elif arr[mid] < search_key:
            low = mid + 1
        else:
            high = mid - 1
    return -1

# Function to measure execution time
def measure_time(func, arr, search_key):
    start_time = time.time()
    func(arr, search_key)
    end_time = time.time()
    return end_time - start_time

# Generate the random array
random_array = generate_random_array()

# Search keys for 5 searches
search_keys = [random.randint(1, 1000) for _ in range(5)]

# Measure the time taken for each search using both algorithms
linear_search_times = []
binary_search_times = []

# Linear search times
for key in search_keys:
    linear_search_times.append(measure_time(linear_search, random_array, key))

# Sort the array for binary search
sorted_array = sorted(random_array)

# Binary search times
for key in search_keys:
    binary_search_times.append(measure_time(binary_search, sorted_array, key))

# Convert times to milliseconds for better readability
linear_search_times_ms = [time_ms * 1000 for time_ms in linear_search_times]
binary_search_times_ms = [time_ms * 1000 for time_ms in binary_search_times]

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(search_keys, linear_search_times_ms, label='Linear Search', marker='o')
plt.plot(search_keys, binary_search_times_ms, label='Binary Search', marker='x')
plt.xlabel('Search Key')
plt.ylabel('Time taken (milliseconds)')
plt.title('Comparing Performance of Linear vs Binary Search')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def string_to_integer(s):
    if not s:  # If the string is empty
        return 0
    else:
        # Convert the first character to integer and multiply by 10^(len(s)-1)
        # then add it to the recursive call on the rest of the string
        return int(s[0]) * (10 ** (len(s) - 1)) + string_to_integer(s[1:])

# Example usage
s = "13531"
result = string_to_integer(s)
print(f"The integer representation of {s} is {result}.")
```

```python
def reverse_string(s):
    # Base case: empty string or a single character string
    if len(s) <= 1:
        return s
    else:
        # Move the first character to the end and recursively reverse the rest
        return reverse_string(s[1:]) + s[0]

# Example u
s = "pots&pans"
reversed_s = reverse_string(s)
print(f"The reverse of '{s}' is '{reversed_s}'.")
```

```python
def is_palindrome(s):
    # Base case: empty string or a single character string is a palindrome
    if len(s) <= 1:
        return True
    # If the first and last characters don't match, it's not a palindrome
    elif s[0] != s[-1]:
        return False
    # If they match, continue checking the substring excluding the first and last characters
    else:
        return is_palindrome(s[1:-1])

# Example usage
strings = ["racecar", "gohangasalamiimalasagnahog", "notapalindrome"]
for s in strings:
    print(f"'{s}' is a palindrome: {is_palindrome(s)}")
```
